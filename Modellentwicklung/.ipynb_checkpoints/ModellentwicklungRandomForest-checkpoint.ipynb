#### Definieren der Pipeline & Feature Engineering

```python
# Definieren der Funktion zum Extrahieren von Zeitmerkmalen
def extract_time_features(df):
    #df['tmsp'] = pd.to_datetime(df['tmsp'])
    df['hour'] = df['tmsp'].dt.hour
    df['minute'] = df['tmsp'].dt.minute
    df['second'] = df['tmsp'].dt.second
    df['weekday'] = df['tmsp'].dt.weekday
    return df[['hour', 'minute', 'second', 'weekday']]
```

```python
# Erstellen der FunctionTransformer
time_transformer = FunctionTransformer(extract_time_features)
```

```python
# Erstellen der Vorverarbeitungsschritte
preprocessor = ColumnTransformer(
    transformers=[
        #('custom', custom_transformer, ['amount', 'PSP', 'card', 'country'])#,
        ('country', OrdinalEncoder(), ['country']),
        ('PSP', OrdinalEncoder(), ['PSP']),
        ('card', OrdinalEncoder(), ['card']),
        ('time', time_transformer, ['tmsp'])
    ],
    remainder='passthrough'
)


# Hinzufügen der Vorverarbeitungsschritte zur Pipeline
steps = [
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))  # Definieren des Klassifikators | Modellgenauigkeit: 0.47064
    #('classifier', SVC(probability=True))  # probability=True, um Wahrscheinlichkeiten zu erhalten
]
# Hinzufügen des FeatureExtractor zur Pipeline
# steps.insert(1, ('feature_extractor', FeatureExtractor()))

# Erstellen der Pipeline
pipeline = Pipeline(steps)
pipeline
```

```python
# Fit und transformieren der Daten mit dem Preprocessor
transformed_array = preprocessor.fit_transform(df_psp)

# Abfragen der Spaltennamen
#transformed_feature_names = preprocessor.named_transformers_['country'].get_feature_names_out(input_features=['country'])
#transformed_feature_names = np.append(transformed_feature_names, preprocessor.named_transformers_['PSP'].get_feature_names_out(input_features=['PSP']))
#transformed_feature_names = np.append(transformed_feature_names, preprocessor.named_transformers_['card'].get_feature_names_out(input_features=['card']))
#transformed_feature_names = np.append(transformed_feature_names, ['hour', 'minute', 'second', 'weekday'])

# Abfragen der Spaltennamen nach der Transformation
transformed_feature_names = ['country','PSP','card',
                             'hour','minute','second','weekday','id','amount', 'success','3d_secured']


# Konvertieren des transformierten Arrays zurück in ein Dataframe
transformed_df = pd.DataFrame(transformed_array, columns=transformed_feature_names)

transformed_df.head()
```

#### Hyperparameter Optimierung

```python
###### Random-Forest
# Definieren der Hyperparametergitter
param_grid = {
    'classifier__n_estimators': [50, 100, 200],
    'classifier__max_depth': [None, 10, 20, 30],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1, 2, 4],
    #'classifier__max_features': ['auto', 'sqrt']
    'classifier__max_features': ['sqrt']
}

# Erstellen der Grid-Search
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='roc_auc', verbose=1, n_jobs=-1, error_score='raise')

# Anpassen der Grid-Search
grid_search.fit(X_train, y_train)
```

```python
# Vorhersagen auf den Testdaten machen
y_pred_best = grid_search.best_estimator_.predict(X_test)

# Wahrscheinlichkeiten für die positive Klasse vorhersagen
y_pred_proba_best = grid_search.best_estimator_.predict_proba(X_test)[:, 1]
print(y_pred_best)
print(y_pred_proba_best)
```

```python
# Berechnen des AUC-Wertes
auc_best = roc_auc_score(y_test, y_pred_proba_best)
print(f'AUC: {auc_best}')
```

```python
# Berechnung der ROC-Kurve
fpr_best, tpr_best, thresholds_best = roc_curve(y_test, y_pred_proba_best)

# Erstellen der Plotly Figur
fig = go.Figure(
    data=[
        go.Scatter(x=fpr_best, y=tpr_best, mode='lines', name='ROC Curve best'),
        go.Scatter(x=[0, 1], y=[0, 1], mode='lines', name='Random Classifier', line=dict(dash='dash'))
    ],
    layout=go.Layout(
        xaxis=dict(title='False Positive Rate'),
        yaxis=dict(title='True Positive Rate'),
        title=f'ROC Curve (AUC = {auc_best:.2f})'
    )
)

# Anzeigen der Figur
fig.update_layout(legend=dict(
    yanchor="top",
    y=0.99,
    xanchor="left",
    x=0.01,
    bgcolor = 'rgba(255,255,255,0.7)'
))
fig.show()
```

#### Shapley-Values



```python
# Benutzerdefinierte Funktion zum Nachahmen der Pipeline-Predict_proba
def pipeline_predict_proba(X):
    if not isinstance(X, pd.DataFrame):
        X = pd.DataFrame(X, columns=X_train.columns)
    return best_pipeline.predict_proba(X)
```

```python
# Sampling der Hintergrunddaten
X_train_sample = shap.sample(X_train, 5000)

# Erstellen eines Explainers mit der benutzerdefinierten Funktion und den gesampelten Hintergrunddaten
explainer = shap.KernelExplainer(pipeline_predict_proba, X_train_sample)
```

```python
# Erstellen eines Explainers mit der benutzerdefinierten Funktion
explainer = shap.KernelExplainer(pipeline_predict_proba, X_train)

#Using 40328 background data samples could cause slower run times. Consider using shap.sample(data, K) or shap.kmeans(data, K) 
#to summarize the background as K samples.
```

```python
# Berechnen der Shapley-Werte für einen bestimmten Datenpunkt (z. B. den ersten Datenpunkt im Testdatensatz)
sample = X_test.iloc[0, :]
shap_values = explainer.shap_values(sample)
```

```python
# Die shap_values-Matrix enthält die Shapley-Werte für jede Klasse separat.
# Um die Shapley-Werte für alle Klassen zusammenzufassen, können Sie den Durchschnitt über die Klassen berechnen:
avg_shap_values = np.mean(shap_values, axis=0)
```

```python
# Embedding Shapley-Werte und Sample-Instanz in Listen
shap_values_embedded = [np.vstack([shap_value]) for shap_value in shap_values]
sample_embedded = pd.DataFrame([sample])

# Anzeigen der Shapley-Werte
shap.summary_plot(shap_values_embedded, sample_embedded)
```
