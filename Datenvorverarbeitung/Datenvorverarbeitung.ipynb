## Vorverarbeitung der Daten  
In diesem notebook werden die Daten aus der Excel-Datei eingelesen, und die Datumsangaben so verändert, dass sie aussehen wie aktuelle Daten.  
Außerdem werden auf der Basis der gegebenen Daten Beispiele extrahiert, um das Hauptprogramm zu testen.

```python
import pandas as pd
from datetime import datetime
```

```python
# Laden der Excel-Datei in ein Pandas Dataframe
df_psp = pd.read_excel('C:/Users/Ibrom/Studium/DLMDWME01/PSP_Jan_Feb_2019.xlsx', sheet_name='Sheet1')

# Umbenennen der ersten Spalte in 'Id'
df_psp = df_psp.rename(columns={df_psp.columns[0]: 'Id'})

# Add 57 month to transfer date from Jan-Feb 2019 to Okt-Nov 2023
df_psp.tmsp = df_psp.tmsp + pd.DateOffset(months=58)
df_psp.reset_index(drop=True)
df_psp.head()
```

```python
df_psp.to_excel("C:/Users/Ibrom/Studium/DLMDWME01/PSP_Nov_Dez_2023.xlsx" , sheet_name='Sheet1')  
```
